```python
df = pd.read_csv("../datasets/weather_cleaned.csv")
```

## Phân bố giữa các cột

### Giữa các cột numerical với nhau

```python
sns.pairplot(
    df[['min', 'max', 'rain', 'cloud', 'humidi', 'wind', 'pressure']], 
    height = 1.3, aspect = 1,
    plot_kws = {'alpha': 0.1}
)
```

```python
ax = sns.heatmap(
    df[['min', 'max', 'rain', 'cloud', 'humidi', 'wind', 'pressure']].corr(),
    annot = True, cmap = 'seismic_r', vmin = -1, vmax = 1
)
ax.set_yticklabels(ax.get_yticklabels(), rotation = 45, va = 'center');
```

**Một số nhận xét về mối quan hệ giữa các biến từ biểu đồ scatter và heatmap độ tương quan**
- Nhiệt độ cao nhất và thấp nhất trong ngày tương đối đồng biến.
- Lượng mưa thấp xảy ra ở hầu hết các mức nhiệt độ nhưng lượng mưa cao tập chung chủ yếu ở mức từ 25°C - 30°C.
- Độ che phủ mây và nhiệt độ nhìn chung chỉ tương quan nhẹ, khi độ che phủ mây tăng thì nhiệt độ giảm.
- Nhiệt độ thấp nhất và độ ẩm dường như không tương quan trong khi nhiệt độ cao nhất có tương quan âm nhẹ
- Tốc độ gió và nhiệt độ dường như không có độ tương quan, khi tốc độ gió tăng thì nhiệt độ có xu hướng tập trung về mức từ 20°C - 30°C.
- Áp suất mực nước biển và nhiệt độ có tương quan âm.
- Lượng mưa và độ che phủ mây tương quan dương nhẹ
- Lượng mưa dường như không có tương quan với tốc độ gió và áp suất mực nước biển
- Lượng mưa có tương quan dương nhẹ với độ ẩm
-  Độ che phủ mây có tương quan dương với độ ẩm ở mức độ tương đối
- 3 yếu tố độ che phủ mây, tốc độ gió và áp suất mực nước biển dường như không có độ tương quan
