```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../datasets/weather_cleaned.csv')
df
```

# Phân bố trong từng cột

## Các cột numerical và categorical tương ứng

### Max và min (Nhiệt độ cao nhất và thấp nhất)

```python
max, frequencies = df['max'].value_counts().index, df['max'].value_counts().values

# Create a histogram with the label frequencies on the x-axis
plt.bar(max, frequencies)

# Add labels and title
plt.xlabel('Nhiệt độ (°C)')
plt.title('Histogram nhiệt độ cao nhất trong ngày')

# Show the plot
plt.show()

# create a dictionary to store the frequency of each label
freq = {'giaret': 0, 'lanh': 0, 'matme': 0, 'amap': 0, 'nong': 0, 'oia': 0}

# iterate over the labels in the dataframe and update the frequency in the dictionary
for label in df['max_bin']:
    freq[label] += 1

# create a list of the labels in the desired order: 'Giá rét', 'Lạnh', 'Mát mẻ', 'Ấm áp', 'Nóng', 'Oi ả'
label_list = ['Giá rét', 'Lạnh', 'Mát mẻ', 'Ấm áp', 'Nóng', 'Oi ả']

# create a list of the frequencies of each label in the order of the label list
freq_list = [freq[label] for label in ['giaret', 'lanh', 'matme', 'amap', 'nong', 'oia']]

# plot the histogram with the label list and frequency list
plt.bar(label_list, freq_list)

# set the title and labels for the plot
plt.title('Phân loại nhiệt độ cao nhất trong ngày')

# annotate the frequency above each column
for i in range(len(freq_list)):
    plt.annotate(str(freq_list[i]), xy=(label_list[i], freq_list[i]), ha='center', va='bottom')

# show the plot
plt.show()
```

### Rain (Lượng mưa)

```python
# create a frequency table of the 'rain' column
freq_table = df['rain'].value_counts().sort_index()

# create a bar chart of the frequency table
plt.bar(freq_table.index, freq_table.values)

# add labels and a title
plt.xlabel('Lượng mưa (mm)')
plt.title('Histogram lượng mưa')

# set the x-axis to be logarithmic
plt.xscale('log')

# display the plot
plt.show()

# create a dictionary to store the frequency of each label
freq = {'khongmua': 0, 'nho': 0, 'vua': 0, 'to': 0, 'ratto': 0}

# iterate over the labels in the dataframe and update the frequency in the dictionary
for label in df['rain_bin']:
    freq[label] += 1

# create a list of the labels in the desired order
label_list = ['Không mưa', 'Nhỏ', 'Vừa', 'To', 'Rất to']

# create a list of the frequencies of each label in the order of the label list
freq_list = [freq[label] for label in ['khongmua', 'nho', 'vua', 'to', 'ratto']]

# plot the histogram with the label list and frequency list
plt.bar(label_list, freq_list)

# set the title and labels for the plot
plt.title('Phân loại lượng mưa')

# annotate the frequency above each column
for i in range(len(freq_list)):
    plt.annotate(str(freq_list[i]), xy=(label_list[i], freq_list[i]), ha='center', va='bottom')

# show the plot
plt.show()
```

Đa số lượng mưa ghi nhận trong ngày ở mức không mưa, nhỏ và vừa, số ngày ghi nhận mưa to và rất to chiếm số rất ít.

## Các cột categorical

```python
# define the order of wind directions
d_order = ['N', 'NNE', 'NE', 'ENE', 'E', 'ESE', 'SE', 'SSE', 'S', 'SSW', 'SW', 'WSW', 'W', 'WNW', 'NW', 'NNW']

# get the wind direction frequencies
max, frequencies = df['wind_d'].value_counts().reindex(d_order).index, df['wind_d'].value_counts().reindex(d_order).values

# create a bar chart with the label frequencies on the x-axis
fig, ax = plt.subplots(figsize=(10,5))
ax.bar(max, frequencies)

# add labels and title
ax.set_title('Histogram hướng gió')

# show the plot
plt.show()
```

- Các cột categorical đều đã được khảo sát phân bố ở phần outlier trong tiền xử lí dữ liệu. Ở đây nhóm chỉ nêu thêm một số nhận xét về cột wind_d (hướng gió)
- Hướng gió chính ghi nhận ở các tỉnh nước ta là Đông Nam, Đông Bắc (từ ENE - Đông Đông Bắc tới SSE - Nam Đông Nam) và Tây Nam (từ WSW - Tây Tây Nam tới SSW - Nam Tây Nam). Tuy nhiên ở từng khu vực hướng gió chủ yếu sẽ khác nhau.
